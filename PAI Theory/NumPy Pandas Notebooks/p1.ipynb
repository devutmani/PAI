```python
# NumPy Attributes

a1 = np.arange(1, 11)
a2 = np.arange(1, 17, dtype=float).reshape(4, 4)
a3 = np.arange(1, 17, dtype=int).reshape(2, 2, 2, 2)
```

```python
print("<-----A1----->")
print(a1)
print("\n<-----A2----->")
print(a2)
print("\n<-----A3----->")
print(a3)
```

```python
# obj.ndim

print(f"Dimensions of a1: {a1.ndim}")
print(f"Dimensions of a2: {a2.ndim}")
print(f"Dimensions of a3: {a3.ndim}")
```

```python
# obj.shape

print(f"Shape of a1: {a1.shape}")
print(f"Shape of a2: {a2.shape}")
print(f"Shape of a3: {a3.shape}")
```

```python
# obj.size --------> tells total no: of elements of array (multiplication of shape parameters --> 2,3 -> 6)

print(f"Size of a1: {a1.size}")
print(f"Size of a2: {a2.size}")
print(f"Size of a3: {a3.size}")
```

```python
# obj.itemsize -----> tells memory size in bytes of single element of array

print(f"ItemSize of a1: {a1.itemsize}")
print(f"ItemSize of a2: {a2.itemsize}")
print(f"ItemSize of a3: {a3.itemsize}")
```

```python
# obj.dtype

print(f"Data Type of a1: {a1.dtype}")
print(f"Data Type of a2: {a2.dtype}")
print(f"Data Type of a3: {a3.dtype}")
```

```python
# astype

print(a1.dtype)
print(a1.astype(np.int32).dtype)
```

```python
# Array Operations

a1 = np.arange(1, 13, dtype=int).reshape(3, 4)
a2 = np.arange(13, 25, dtype=int).reshape(3,4)
a3 = np.arange(1, 13, 2, dtype=int).reshape(2, 3)

print(a1)
print()

print(a2)
print()

print(a3)
```

```python
# Scalar Operations / Arithmetic Operations

a1 = a1 + 5
print(a1.astype(np.int64))

a1 = a1 - 2
print()
print(a1.astype(np.int64))

a1 = a1 * 3                 # Scalar Multiplication
print()
print(a1.astype(np.int64))

a1 = a1 / 4
print()
print(a1.astype(np.int64))

a1 = a1 % 6
print()
print(a1.astype(np.int64))

a1 = a1 ** 3
print()
print(a1.astype(np.int64))
```

```python
# Relational Operations

a1 == 0
a1 != 0
a1 > 5
a1 < 5
a1 >= 5
a1 <= 5
```

```python
# Vector Operations

a3 = a1+a2
print()
print(a3.astype(np.int64))

a3 = a1-a2
print()
print(a3.astype(np.int64))

a3 = a1*a2
print()
print(a3.astype(np.int64))

a3 = a1/a2
print()
print(a3.astype(np.int64))

a3 = a1**a2
print()
print(a3)

a3 = a1%a2
print()
print(a3.astype(np.int64))
```

```python
# Array Functions

a1 = np.random.random((3, 4))
a1 = np.round(a1*100)
print(a1)
```

```python
# max/min/sum/prod

print(f"Max of Vector: {np.max(a1)}")
print(f"Min of Vector: {np.min(a1)}")
print(f"Sum of Vector: {np.sum(a1)}")
print(f"Prod of Vector: {np.prod(a1)}")
```

```python
# Initalizing Matrix

a = np.arange(1, 41, 2, dtype=int).reshape(4, 5)
print(a)
```

```python
# array of min's of each row
b = np.min(a, axis=1)
print(b)

# array of min's of each colm
c = np.min(a, axis=0)
print(c)
```

```python
# array of max's of each row
b = np.max(a, axis=1)
print(b)

# array of max's of each colm
c = np.max(a, axis=0)
print(c)
```

```python
# array of prod of each row
b = np.prod(a, axis=1)
print(b)

# array of prod of each colm
c = np.prod(a, axis=0)
print(c)
```

```python
# array of sum of each row
b = np.sum(a, axis=1)
print(b)

# array of sum of each colm
c = np.sum(a, axis=0)
print(c)
```

```python
print(f"Mean of a: {np.mean(a)}")
print(f"Median of a: {np.median(a)}")
print(f"Standard Deviation of a: {np.std(a)}")
print(f"Variance of a: {np.var(a)}")
```

```python
# array of Mean of each row
b = np.mean(a, axis=1)
print(b)

# array of Mean of each colm
b = np.mean(a, axis=0)
print(b)
```

```python
# array of Median of each row
b = np.median(a, axis=1)
print(b)

# array of Median of each colm
b = np.median(a, axis=0)
print(b)
```

```python
# array of Std of each row
b = np.std(a, axis=1)
print(b)

# array of Std of each colm
b = np.std(a, axis=0)
print(b)
```

```python
# array of Variance of each row
b = np.var(a, axis=1)
print(b)

# array of Variance of each colm
b = np.var(a, axis=0)
print(b)
```

```python
# dot Product

a = np.arange(1, 11).reshape(2, 5)
b = np.arange(11, 21).reshape(5, 2)

np.dot(a, b)
```

```python
# log and exponent 

np.log(a)   # log of each element

np.exp(a)   # exponent of each element
```

```python
# round, floor and ceil

a = np.random.random((2, 3))
print(a, "\n")

a = np.round(a*100)
print(a, "\n")

a = np.round(a)
print(a, "\n")

a = np.floor(a)
print(a, "\n")

a = np.ceil(a)
print(a, "\n")
```

```python
# Indexing and Slicing

a1 = np.arange(10, dtype=int)
a2 = np.arange(12, dtype=int).reshape(3, 4)

print(a1)
print("\n", a2)
```

```python
print(a1[2::2], "\n")
print(a2[:2,1:], "\n")
```
